```python
import pandas as pd
```

```python
ukb_df = pd.read_csv("../data/sex_lf_bias/ukb_sex_lf_bias.csv.gz")
aou_df = pd.read_csv("../data/sex_lf_bias/aou_sex_lf_bias.csv.gz")
```

```python
ukb_df.loc[ukb_df.sedentary_pvalue<0.05].sort_values("sedentary_pvalue")
```

```python
meta_df = ukb_df.merge(aou_df, on="ID", suffixes=("_ukb", "_aou"))
```

```python
sig_df = pd.DataFrame()

for lf in ["genetic_sex", "pa", "alcohol", "smoke"]:
    lf_df = meta_df.loc[(meta_df[f"{lf}_pvalue_ukb"]<0.05)&(meta_df[f"{lf}_pvalue_aou"]<0.05)]
    # get the most significant one
    if len(lf_df)>0:
        lf_df = lf_df.loc[lf_df.groupby(lf_df.ID.str.split(".", expand=True).iloc[:, 0])[f"{lf}_pvalue_ukb"].idxmin()]
        lf_df["lifestyle"] = lf
        sig_df = pd.concat((lf_df, sig_df))
```

```python
sig_df.reset_index(drop=True).to_csv("../data/sex_lf_bias/sex_lf_bias.csv.gz", index=False)
```
